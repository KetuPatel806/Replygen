```python
import json, datetime, asyncio
from dataclasses import dataclass, field
from langgraph.graph import StateGraph, START, END
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
```

```python
## LLM Initialization
llm=ChatGroq(model="llama-3.3-70b-versatile")
llm.bind_tools(tools)

llm.invoke("Hey there")
```

```python
## DataClasses
@dataclass
class AgentState:
    platform: str
    post_text: str
    tone: str | None = None
    intent: str | None = None
    draft: str | None = None
    reply: str | None = None
    ts: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
```

```python
## DetectTone Node
def DetectTone(state: AgentState) -> AgentState:
    system_msg = (
        "You are a toneâ€‘classification microservice.\n"
        "â€¢ Task: Read the userâ€™s socialâ€‘media post and decide the single bestâ€‘fit tone.\n"
        "â€¢ Allowed output: EXACTLY ONE lowerâ€‘case word, no punctuation, no explanations.\n"
        "â€¢ Legal tone vocabulary (choose the closest match):\n"
        "  enthusiastic | celebratory | casual | humorous | professional | authoritative | critical | urgent | contemplative | grateful | humble | inspirational | formal | sarcastic | negative\n"
        "â€¢ If none fit perfectly, choose the nearest.\n"
        "â€¢ Do NOT output JSON, numbers, multiple words, or anything besides the tone word.\n"
        "Respond with that word ONLY."
    )
    res = llm.invoke([
        ("system", system_msg),
        ("human", state.post_text)
    ])
    state.tone = res.content.strip().split()[0].lower()
    return state
```

```python
## Extract Intent Node
def ExtractIntent(state: AgentState) -> AgentState:
    system_msg = (
        "You are an intentâ€‘extraction microservice.\n"
        "â€¢ Goal: Summarize the PRIMARY intent or purpose of the userâ€™s post in â‰¤â€¯15 words.\n"
        "â€¢ Output: ONE sentence, no bullet points, no hashtags, no emojis.\n"
        "Focus on what the poster is trying to achieve (announce news, ask for help, share opinion, etc.)."
    )
    res = llm.invoke([("system", system_msg), ("human", state.post_text)])
    state.intent = res.content.strip()
    return state
```

```python
## DraftReply Node
def DraftReply(state: AgentState) -> AgentState:
    system_msg = (
        f"You are an avid, authentic {state.platform} user.\n"
        f"Craft a firstâ€‘draft reply that:\n"
        f"â€¢ Aligns with the detected tone: {state.tone}.\n"
        f"â€¢ Directly addresses the posterâ€™s intent: {state.intent}.\n"
        f"â€¢ Feels human (no â€œAs an AIâ€¦â€ or overly formal phrases).\n"
        f"â€¢ Length: 20â€‘40 words for Twitter, 30â€‘70 words for LinkedIn, 10â€‘30 words for Instagram.\n"
        f"â€¢ Twitter: â‰¤ 1 tasteful emoji + no links.  LinkedIn: professional but warm, no emojis.  "
        f"Instagram: emojis allowed, keep it light.\n"
        f"Reply ONLYâ€”donâ€™t restate the original post."
    )
    res = llm.invoke([("system", system_msg), ("human", state.post_text)])
    state.draft = res.content.strip()
    return state
```

```python
## Refine Reply Node
def RefineReply(state: AgentState) -> AgentState:
    system_msg = (
        f"You are an editor polishing a socialâ€‘media reply for {state.platform}.\n"
        f"Constraints:\n"
        f"â€¢ Preserve original meaning and tone ({state.tone}).\n"
        f"â€¢ Tighten wording; max 50 words overall.\n"
        f"â€¢ Remove fluff, clichÃ©s, AI giveaways, or duplicate content.\n"
        f"â€¢ Respect platform norms (see below):\n"
        f"    â€“ Twitter: 1â€‘2 emojis allowed, no hashtags unless organic.\n"
        f"    â€“ LinkedIn: no emojis, keep it professional yet friendly.\n"
        f"    â€“ Instagram: emojis and light exclamation OK, no links.\n"
        f"Return **only the final reply text**."
    )
    res = llm.invoke([("system", system_msg), ("human", state.draft)])
    state.reply = res.content.strip()
    return state
```

```python
## StateGraph
graph = StateGraph(AgentState)
graph.add_node("DetectTone", DetectTone)
graph.add_node("ExtractIntent", ExtractIntent)
graph.add_node("DraftReply", DraftReply)
graph.add_node("RefineReply", RefineReply)
graph.add_edge(START, "DetectTone")
graph.add_edge("DetectTone", "ExtractIntent")
graph.add_edge("ExtractIntent", "DraftReply")
graph.add_edge("DraftReply", "RefineReply")
graph.add_edge("RefineReply", END)
pipeline = graph.compile()
```

```python
## invoking the Pipeline
async def run_example():
    init = AgentState(
        platform="Twitter",
        post_text="Thrilled to announce our new AIâ€‘powered product launch next week!"
    )
    result = await pipeline.ainvoke(init)
    print("Final reply:", result["reply"])
    print("\nFull state:", result)


await run_example()
```
